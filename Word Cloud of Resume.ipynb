```python
import PyPDF2 #Library to access PDF files.
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Open the PDF file and save it as a variable
resume = open('resume.pdf', 'rb')
```

<p>Syntax : open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)<br>
'r'	Open a file for reading. (default)<br>
'w'	Open a file for writing. Creates a new file if it does not exist or truncates the file if it exists.<br>
'x'	Open a file for exclusive creation. If the file already exists, the operation fails.<br>
'a'	Open for appending at the end of the file without truncating it. Creates a new file if it does not exist.<br>
't'	Open in text mode. (default)<br>
'b'	Open in binary mode.<br>
'+'	Open a file for updating (reading and writing)</p>

```python
resume
```

```python
# Creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(resume) 
```

```python
# Find the Number of pages in the pdfReader object
print(pdfReader.numPages) 
```

```python
# Create a page object for getting the text of the pdf file
pageObj = pdfReader.getPage(0)
```

```python
# Print the text present in the PDF file
resume_text = pageObj.extractText()
resume_text
```

```python
# Close the opened PDF file
resume.close()
```

```python
# Removing the '\n' characters
resume_text=resume_text.replace('\n',"")
resume_text
```

```python
# Prepare a 800x800 square word cloud 
resume_wordcloud =  WordCloud(width = 800, height = 800, 
                    background_color ='white',
                    min_font_size = 10).generate(resume_text) 
resume_wordcloud
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
stopwords = set(STOPWORDS)
stopwords
```

```python
# Prepare New Word CLud without Stop Words
resume_wordcloud =  WordCloud(width = 800, height = 800, 
                    background_color ='white',
                    stopwords = stopwords,      
                    min_font_size = 10).generate(resume_text) 
resume_wordcloud
```

```python
# plot the WordCloud image without stop words                       
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
irrelevant_words = ['Aug','using','Pvt','July','India']
```

```python
# Remove irrelevant Words from the resume_text
for word in irrelevant_words:
    if word in resume_text:
        resume_text=resume_text.replace(word,"")
resume_text
```

```python
# Print the filtered Square word cloud
resume_wordcloud =  WordCloud(width = 600, height = 600, 
                    background_color ='white',
                    stopwords = stopwords,      
                    min_font_size = 10).generate(resume_text) 
resume_wordcloud

# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
# Open the image over which we need to write our Word Cloud
mask = np.array(Image.open("Smiley.png"))
```

```python
wordcloud_resume_text = WordCloud(stopwords=stopwords, 
                                  background_color="white", 
                                  mode="RGBA", 
                                  mask=mask).generate(resume_text)
```

```python
# Disply the word cloud
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[9,9])
plt.imshow(wordcloud_resume_text.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```

```python
# Save the Smiley word cloud as a PNG impage in current folder
wordcloud_resume_text.to_file("resume_wordcloud.png")
```
